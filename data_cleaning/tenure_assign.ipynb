```python
# TENURE ASSIGNMENT - START!

# Where we left:
# ├─ Separate groups (unemployed + employed)
# ├─ Each credit tier
# ├─ 5 tenures: [12, 18, 24, 30, 36]
# ├─ 20% each distribution
# └─ Ready to code!

# FIRST: Define all groups

# Code:

# UNEMPLOYED GROUPS
unemp_worst = (df['employed'] == 0) & (df['credit_score'] <= 400)
unemp_poor = (df['employed'] == 0) & (df['credit_score'].between(401, 550))
unemp_good = (df['employed'] == 0) & (df['credit_score'].between(551, 699))
unemp_best = (df['employed'] == 0) & (df['credit_score'] >= 700)

# EMPLOYED GROUPS
emp_worst = (df['employed'] == 1) & (df['credit_score'] <= 400)
emp_poor = (df['employed'] == 1) & (df['credit_score'].between(401, 550))
emp_good = (df['employed'] == 1) & (df['credit_score'].between(551, 699))
emp_best = (df['employed'] == 1) & (df['credit_score'] >= 700)

print(f"Unemployed worst: {unemp_worst.sum()}")
print(f"Unemployed poor: {unemp_poor.sum()}")
print(f"Unemployed good: {unemp_good.sum()}")
print(f"Unemployed best: {unemp_best.sum()}")
print("\n\n")
print(f"Employed worst: {emp_worst.sum()}")
print(f"Employed poor: {emp_poor.sum()}")
print(f"Employed good: {emp_good.sum()}")
print(f"Employed best: {emp_best.sum()}")

print(unemp_worst.sum()+ unemp_poor.sum()+ unemp_good.sum()+unemp_best.sum()+emp_worst.sum()+emp_poor.sum()+emp_good.sum()+emp_best.sum())
# ... etc
```

```python
# UNEMPLOYED WORST
unemp_worst = (df['employed'] == 0) & (df['credit_score'] <= 400)
n = unemp_worst.sum()
indices = df[unemp_worst].index
df.loc[indices[:n//5], 'loan_tenure_months'] = 12
df.loc[indices[n//5:2*n//5], 'loan_tenure_months'] = 18
df.loc[indices[2*n//5:3*n//5], 'loan_tenure_months'] = 24
df.loc[indices[3*n//5:4*n//5], 'loan_tenure_months'] = 30
df.loc[indices[4*n//5:], 'loan_tenure_months'] = 36

# UNEMPLOYED POOR
unemp_poor = (df['employed'] == 0) & (df['credit_score'].between(401, 550))
n = unemp_poor.sum()
indices = df[unemp_poor].index
df.loc[indices[:n//5], 'loan_tenure_months'] = 12
df.loc[indices[n//5:2*n//5], 'loan_tenure_months'] = 18
df.loc[indices[2*n//5:3*n//5], 'loan_tenure_months'] = 24
df.loc[indices[3*n//5:4*n//5], 'loan_tenure_months'] = 30
df.loc[indices[4*n//5:], 'loan_tenure_months'] = 36

# UNEMPLOYED GOOD
unemp_good = (df['employed'] == 0) & (df['credit_score'].between(551, 699))
n = unemp_good.sum()
indices = df[unemp_good].index
df.loc[indices[:n//5], 'loan_tenure_months'] = 12
df.loc[indices[n//5:2*n//5], 'loan_tenure_months'] = 18
df.loc[indices[2*n//5:3*n//5], 'loan_tenure_months'] = 24
df.loc[indices[3*n//5:4*n//5], 'loan_tenure_months'] = 30
df.loc[indices[4*n//5:], 'loan_tenure_months'] = 36

# UNEMPLOYED BEST
unemp_best = (df['employed'] == 0) & (df['credit_score'] >= 700)
n = unemp_best.sum()
indices = df[unemp_best].index
df.loc[indices[:n//5], 'loan_tenure_months'] = 12
df.loc[indices[n//5:2*n//5], 'loan_tenure_months'] = 18
df.loc[indices[2*n//5:3*n//5], 'loan_tenure_months'] = 24
df.loc[indices[3*n//5:4*n//5], 'loan_tenure_months'] = 30
df.loc[indices[4*n//5:], 'loan_tenure_months'] = 36

print("Unemployed tenure assigned!")
```
