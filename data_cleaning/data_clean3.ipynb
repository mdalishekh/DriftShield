```python
import pandas as pd
import numpy as np
FILE = "csv-2/copy.csv"
df = pd.read_csv(FILE)
```

```python
df.head()
```

## Assigning loan to UN Employed

```python
## POOR Cibil

poor_unemp = (df['employed'] == 0) & (df['credit_score'].between(401, 550))
```

```python


df.loc[(poor_unemp) & (df['age'].between(25, 29)), 'loan_amount'] = \
    np.random.randint(40000, 70000, size=df[(poor_unemp) & (df['age'].between(25, 29))].shape[0])

df.loc[(poor_unemp) & (df['age'].between(30, 40)), 'loan_amount'] = \
    np.random.randint(50000, 100000, size=df[(poor_unemp) & (df['age'].between(30, 40))].shape[0])

df.loc[(poor_unemp) & (df['age'].between(41, 55)), 'loan_amount'] = \
    np.random.randint(70000, 100000, size=df[(poor_unemp) & (df['age'].between(41, 55))].shape[0])

df.loc[(poor_unemp) & (df['age'].between(56, 59)), 'loan_amount'] = \
    np.random.randint(50000, 80000, size=df[(poor_unemp) & (df['age'].between(56, 59))].shape[0])

df.loc[(poor_unemp) & (df['age'] >= 60), 'loan_amount'] = \
    np.random.randint(40000, 70000, size=df[(poor_unemp) & (df['age'] >= 60)].shape[0])

print(f"Unemployed poor credit loans assigned! {poor_unemp.sum()} rows")
```

```python
df[poor_unemp]
```

```python
df.loc[
    (df["loan_amount"]>0)
]
```
