```python
# import pandas as pd
# print("HI there")
# from evidently import Report 
# from evidently.presets import DataDriftPreset


# reference_data = pd.DataFrame({
#     "age": [20, 25, 30, 35, 40],
#     "income": [20000, 30000, 40000, 50000, 60000]
# })

# current_data = pd.DataFrame({
#     "age": [50, 55, 60, 65, 70],
#     "income": [100000, 120000, 140000, 160000, 180000]
# })


# report = Report(
#     metrics=[
#         DataDriftPreset()
#     ]
# )

# report.run(
#     reference_data=reference_data,
#     current_data=current_data
# )


# report.save_html(
#     "drift_report.html"
# )

# print("HTML generated successfully")

import json
import pandas as pd

from evidently import Report
from evidently.presets import DataDriftPreset
from evidently.metrics import DriftedColumnsCount


# Reference Dataset
reference_data = pd.DataFrame({
    "age": [20, 25, 30, 35, 40],
    "income": [20000, 30000, 40000, 50000, 60000]
})

# Current Dataset
current_data = pd.DataFrame({
    "age": [50, 55, 60, 65, 70],
    "income": [100000, 120000, 140000, 160000, 180000]
})


# Create Report

report = Report(
    metrics=[
        DataDriftPreset()
    ]
)


# report = Report(
#     metrics=[
#         DriftedColumnsCount()
#     ]
# )

# Run Report
snapshot = report.run(
    reference_data=reference_data,
    current_data=current_data
)


# ==========================================
# SAVE HTML
# ==========================================

snapshot.save_html(
    "drift_report.html"
)

print("HTML Saved Successfully")


# ==========================================
# SAVE JSON
# ==========================================

snapshot.save_json(
    "drift_report.json"
)

print("JSON Saved Successfully")


# ==========================================
# DICT OUTPUT
# ==========================================

data = snapshot.dict()

print("\nType:")
print(type(data))

print("\nKeys:")
print(data.keys())


# ==========================================
# PRETTY PRINT FIRST PART
# ==========================================

print("\nFirst 5000 chars:\n")

print(
    json.dumps(
        data,
        indent=2,
        default=str
    )[:5000]
)

print("\n\n\n\n-----------------------\n\n\n")

data = snapshot.dict()

for metric in data["metrics"]:

    print("\n------------------")

    print(metric["metric_name"])

    print(metric["value"])
    
print("\n\n\n")    
snapshot.metric_results
```
