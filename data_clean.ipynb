```python
import pandas as pd

df = pd.read_csv('csv/loan_data.csv')

df.head(10)
```

```python
# Check missing values

print(df.isnull().sum())

# Data types
print(df.dtypes)
```

```python
percent_default = lambda a, b: (a/b)*100

(df["default"].value_counts()[0] / df["default"].count()) *100
```

```python
total_defaults = df["default"].value_counts()[0]
total_counts = df["default"].count()
print(total_defaults, total_counts)

print(percent_default(total_defaults, total_counts))

# 79.2 % Loan default hai 
# 100 - 79.2 = 20.8 % logon ne loan repay kiya hai
total_non_defaults = df["default"].value_counts()[1]
print(percent_default(total_non_defaults, total_counts))
```
